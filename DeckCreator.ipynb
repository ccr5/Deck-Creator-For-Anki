```python
import pandas as pd
import requests
import bs4
```

```python
# Create a function to get all descriptions
def GetDescription(response):
    soup = bs4.BeautifulSoup(response.text, 'lxml')
    return [x.text for x in soup.select('.quick-results.container')[0].select('.quick-result-entry .quick-result-overview .sense-group-results a')]
```

```python
# Test GetDescription()
word = 'play'
res = requests.get(f'https://pt.bab.la/dicionario/ingles-portugues/{word}')
desc = GetDescription(res)
desc
```

```python
# Create a function to get all examples
def GetSamples(response):

    soup = bs4.BeautifulSoup(response.text, 'lxml')
    example_list = []

    for example in [
            x.text.split()
            for x in soup.select('.result-block.container .sense-group .dict-entry .more-examples .dict-source')]:
        count = 0
        ret = []
        for x in example:
            if x == 'revisão':
                count = 1
                continue

            if count == 0:
                pass
            else:
                ret.append(
                    x.replace("'", "").replace('"', '').replace(';', '.')
                )
        example_list.append(ret)
        
    return example_list
```

```python
# Test GetSamples()
res = requests.get(f'https://pt.bab.la/dicionario/ingles-portugues/{word}')
samples = GetSamples(res)
samples
```
